```python
import random

list = ['S','W','G']

computer_count = 0
player_count = 0
count = 0

while (count != 10):
    player_choice = input("Give an input - S for Snake, W for Water, G for Gun : ")
    print("the count is ",count)
    print("the player choise is ",player_choice)
    if player_choice.upper() in list:
        computer_choice = random.choice(list)
        
        print("the computer choise is ",computer_choice)
        count += 1
        
        if player_choice == computer_choice:
            print('No One Won!!!')
        
        elif computer_choice == 'S'and player_choice == 'W':
            print(f'Computer has WON with choice {computer_choice}')
            computer_count += 1
        
        elif computer_choice == 'W'and player_choice == 'G':
            print(f'Computer has WON with choice {computer_choice}')
            computer_count += 1
        
        elif computer_choice == 'G'and player_choice == 'S':
            print(f'Computer has WON with choice {computer_choice}')
            computer_count += 1
        
        else:
            print(f'Player has WON and computer input was {computer_choice}')
            player_count += 1
    
    else:
        print('You have entered wrong input!!!!',player_choice)
        
else:
    if (computer_count == player_count):
        print('Its a Draw')
        
    elif (computer_count > player_count):
        print(f'Computer has won game with {computer_count}')
        
    else:
        print(f'Player has won game with {player_count}')
    
```
